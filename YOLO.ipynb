```python
import numpy as np
from PIL import Image
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import cv2

# Load the YOLO model
model = YOLO("yolov8n.pt")

# Load the image
image = Image.open("cv.png")

# Convert image to OpenCV format (BGR)
image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Detect objects in the image
results = model.predict(image)

# Print detection results
print(results)

# Initialize list of colors for bounding boxes
liste_de_couleur = [(0, 255, 0), (255, 255, 0), (0, 255, 255), (255, 0, 0), (255, 255, 255), (0, 0, 255)]
i = 0

# Loop over the detected objects
for data in results[0].boxes.data:
    x, y, x2, y2 = data[0], data[1], data[2], data[3]
    confidence = data[4] # Probability
    label_index = int(data[5]) # Class index

    if confidence > 0.3:  # Set your confidence threshold here
        print(confidence)

        # Check if the detected object is a person
        if label_index == 0:  # Assuming 'person' class is at index 0
            # Draw a rectangle around the object
            cv2.rectangle(image_cv, (int(x), int(y)), (int(x2), int(y2)), liste_de_couleur[i], 2)
            print("Person detected with confidence:", confidence, i)

            # Crop the bounding box for the person
            person_crop = image_cv[int(y):int(y2), int(x):int(x2)]

            # Save the cropped image
            cv2.imwrite("person_crop2.png", person_crop)

        i = i + 1
        if i > 5:
            i = 0

# Display the image with the rectangles
cv2_imshow(image_cv)
```

```python
import os
import numpy as np
from PIL import Image
from ultralytics import YOLO
import cv2

# Load the YOLO model
model = YOLO("yolov8n.pt")

# Define input and output directories
input_dir = "path/to/your/input/folder"
output_dir = "path/to/your/output/folder"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Initialize list of colors for bounding boxes
liste_de_couleur = [(0, 255, 0), (255, 255, 0), (0, 255, 255), (255, 0, 0), (255, 255, 255), (0, 0, 255)]

# Loop over all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):  # Process only image files
        image_path = os.path.join(input_dir, filename)

        # Load the image
        image = Image.open(image_path)

        # Convert image to OpenCV format (BGR)
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

        # Detect objects in the image
        results = model.predict(image)

        # Print detection results
        print(results)

        i = 0

        # Loop over the detected objects
        for data in results[0].boxes.data:
            x, y, x2, y2 = data[0], data[1], data[2], data[3]
            confidence = data[4]  # Probability
            label_index = int(data[5])  # Class index

            if confidence > 0.3:  # Set your confidence threshold here
                print(confidence)

                # Check if the detected object is a person
                if label_index == 0:  # Assuming 'person' class is at index 0
                    # Draw a rectangle around the object
                    cv2.rectangle(image_cv, (int(x), int(y)), (int(x2), int(y2)), liste_de_couleur[i], 2)
                    print("Person detected with confidence:", confidence, i)

                    # Crop the bounding box for the person
                    person_crop = image_cv[int(y):int(y2), int(x):int(x2)]

                    # Save the cropped image
                    output_path = os.path.join(output_dir, filename)
                    cv2.imwrite(output_path, person_crop)

                i = i + 1
                if i > 5:
                    i = 0

        # Optionally display the image with the rectangles (for debugging purposes)
        # cv2.imshow("Detected Image", image_cv)
        # cv2.waitKey(0)
        # cv2.destroyAllWindows()
```
